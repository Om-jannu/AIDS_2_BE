```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0, stratify = y)
```

```python
#constant feature removal

constant_filter = VarianceThreshold(threshold=0)
constant_filter.fit(x_train)
```

```python
constant_filter.get_support().sum()
constant_list = [not temp for temp in constant_filter.get_support()]
constant_list
```

```python
x_train_filter = constant_filter.transform(x_train)
x_test_filter = constant_filter.transform(x_test)
x_train_filter.shape,x_test_filter.shape,x_train.shape
```

```python
#quasi constant removal

quasi_constant_filter = VarianceThreshold(threshold=0.1)
```

```python
quasi_constant_filter.fit(x_train_filter)
quasi_constant_filter.get_support().sum()
```

```python
x_train_quasi_filter = quasi_constant_filter.transform(x_train_filter)
x_test_quasi_filter = quasi_constant_filter.transform(x_test_filter)
x_train_quasi_filter.shape, x_test_quasi_filter.shape
```

```python
#remove duplicate rows

x_train = x_train_quasi_filter.T
x_test = x_test_quasi_filter.T
```

```python
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
x_train.duplicated().sum()
```

```python
duplicated_features = x_train.duplicated()
feature_to_keep = [not index for index in duplicated_features]
```

```python
x_train_unique = x_train[feature_to_keep].T
x_test_unique = x_test[feature_to_keep].T
x_train_unique.shape, x_test_unique.shape
```

```python
def run_randomF(x_train, x_test, y_train, y_test):
  clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
  clf.fit(x_train, y_train)
  y_pred = clf.predict(x_test)
  print('Accuracy on test: ')
  print(accuracy_score(y_test, y_pred))
```

```python
%%time
run_randomF(x_train_unique, x_test_unique, y_train, y_test)
```

```python
def get_correlation(data, threshold):
  corr_col = set()
  cor_rel_mat = data.corr()
  for i in range (len(cor_rel_mat.columns)):
    for j in range (i):
      if abs(cor_rel_mat.iloc[i,j])>threshold:
        colname = cor_rel_mat.columns[i]
        corr_col.add(colname)
  return corr_col
```

```python
corr_features = get_correlation(x_train_unique, 0.85)
corr_features
```

```python
len(corr_features)
```

```python
x_train_uncorr = x_train_unique.drop(labels = corr_features, axis = 1)
x_test_uncorr = x_test_unique.drop(labels = corr_features, axis = 1)
x_train_uncorr.shape, x_test_uncorr.shape
```

```python
%%time
run_randomF(x_train_uncorr, x_test_uncorr, y_train, y_test)
```
